## Logistic Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets as datasets
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures

centers = [[0, 0]]
t, _ = datasets.make_blobs(n_samples=750, centers=centers, cluster_std=1, random_state=0)

# LINE
def generate_line_data():
    # create some space between the classes
    X = np.array(list(filter(lambda x : x[0] - x[1] < -.5 or x[0] - x[1] > .5, t)))
    Y = np.array([1 if x[0] - x[1] >= 0 else 0 for x in X])
    return X, Y

# CIRCLE
def generate_circle_data(t):
    # create some space between the classes
    X = np.array(list(filter(lambda x : (x[0] - centers[0][0])**2 + (x[1] - centers[0][1])**2 < 1 or (x[0] - centers[0][0])**2 + (x[1] - centers[0][1])**2 > 1.5, t)))
    Y = np.array([1 if (x[0] - centers[0][0])**2 + (x[1] - centers[0][1])**2 >= 1 else 0 for x in X])
    return X, Y

# XOR
def generate_xor_data():
    X = np.array([
        [0,0],
        [0,1],
        [1,0],
        [1,1]])
    Y = np.array([x[0]^x[1] for x in X])
    return X, Y
```

a) Using the above code, generate and plot data that is linearly separable.

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets as datasets

centers = [[0, 0]]
t, _ = datasets.make_blobs(n_samples=750, centers=centers, cluster_std=1, random_state=0)


def generate_line_data():
    X = np.array(list(filter(lambda x : x[0] - x[1] < -.5 or x[0] - x[1] > .5, t)))
    Y = np.array([1 if x[0] - x[1] >= 0 else 0 for x in X])
    return X, Y

X_line, Y_line = generate_line_data()

plt.figure(figsize=(8, 6))
plt.scatter(X_line[:, 0], X_line[:, 1], c=Y_line, cmap='bwr', edgecolors='k')
plt.title("Linearly Separable Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

b) Fit a logistic regression model to the data a print out the coefficients.

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression().fit(X_line, Y_line)


coefficients = model.coef_
intercept = model.intercept_

coefficients, intercept
```

c) Using the coefficients, plot the line through the scatter plot you created in a). (Note: you need to do some math to get the line in the right form)

```python
slope = -coefficients[0,0] / coefficients[0,1]
intercept_line = -intercept / coefficients[0,1]


x_values = np.linspace(X_line[:, 0].min(), X_line[:, 0].max(), 300)

y_values = slope * x_values + intercept_line

plt.figure(figsize=(8, 6))
plt.scatter(X_line[:, 0], X_line[:, 1], c=Y_line, cmap='bwr', edgecolors='k')
plt.plot(x_values, y_values, 'k--')  # Plotting the decision boundary as a dashed line
plt.title("Linearly Separable Data with Decision Boundary")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

d) Using the above code, generate and plot the CIRCLE data.

```python
def generate_circle_data(t):
    X = np.array(list(filter(lambda x : (x[0] - centers[0][0])**2 + (x[1] - centers[0][1])**2 < 1 or (x[0] - centers[0][0])**2 + (x[1] - centers[0][1])**2 > 1.5, t)))
    Y = np.array([1 if (x[0] - centers[0][0])**2 + (x[1] - centers[0][1])**2 >= 1 else 0 for x in X])
    return X, Y

X_circle, Y_circle = generate_circle_data(t)

# Plotting the circle data
plt.figure(figsize=(8, 6))
plt.scatter(X_circle[:, 0], X_circle[:, 1], c=Y_circle, cmap='bwr', edgecolors='k')
plt.title("Circle Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

g) Plot the XOR data. In this 2D space, the data is not linearly separable, but by introducing a new feature $$x_3 = x_1 * x_2$$

(called an interaction term) we should be able to find a hyperplane that separates the data in 3D. Plot this new dataset in 3D.

```python
from mpl_toolkits.mplot3d import Axes3D

# Generate XOR data
def generate_xor_data():
    X = np.array([
        [0,0],
        [0,1],
        [1,0],
        [1,1]])
    Y = np.array([x[0]^x[1] for x in X])
    return X, Y

X_xor, Y_xor = generate_xor_data()

# Create 3D plot with interaction term x3 = x1 * x2
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(X_xor[:, 0], X_xor[:, 1], X_xor[:, 0] * X_xor[:, 1], c=Y_xor, cmap='bwr', depthshade=True)
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X1 * X2')
plt.show()
```

h) Apply a logistic regression model using the interaction term. Plot the decision boundary.

```python

poly_xor = PolynomialFeatures(interaction_only=True, include_bias=False)
lr_xor = LogisticRegression()
model_xor = make_pipeline(poly_xor, lr_xor).fit(X_xor, Y_xor)


x_min_xor, x_max_xor = X_xor[:, 0].min() - .5, X_xor[:, 0].max() + .5
y_min_xor, y_max_xor = X_xor[:, 1].min() - .5, X_xor[:, 1].max() + .5
xx_xor, yy_xor = np.meshgrid(np.arange(x_min_xor, x_max_xor, h),
                             np.arange(y_min_xor, y_max_xor, h))

meshData_xor = np.c_[xx_xor.ravel(), yy_xor.ravel()]

A_xor = model_xor.predict_proba(meshData_xor)[:, 1].reshape(xx_xor.shape)
Z_xor = model_xor.predict(meshData_xor).reshape(xx_xor.shape)

fig, ax = plt.subplots()
contour = ax.contourf(xx_xor, yy_xor, A_xor, cmap="RdBu", vmin=0, vmax=1)
ax.axis('off')

colors = ['red' if y else 'blue' for y in Y_xor]
ax.scatter(X_xor[:, 0], X_xor[:, 1], color=colors, s=50, alpha=0.9)

fig.colorbar(contour, ax=ax, label='Probability of class 1')
plt.show()
```
