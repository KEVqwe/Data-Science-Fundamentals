a) Create and plot a dataset of 100 (Y, X) points where `Y = 1 + 0.5 * X + eps`.

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

SAMPLE_SIZE = 100

# Defining the true coefficients
true_beta = np.array([1, 0.5])

# Generating the independent variable data
x = -1.0 + 2.0 * np.random.random(SAMPLE_SIZE)

# Generating the dependent variable data with added noise
y = true_beta[0] + true_beta[1] * x + np.random.randn(SAMPLE_SIZE)

# Plotting the data
plt.plot(x, y, 'ro', markersize=4)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Dataset of Y = 1 + 0.5X + eps')
plt.grid(True)
plt.show()
```

b) Plot the least squares estimate line through the scatter plot.

```python
# Computing the least squares estimate for the coefficients
intercept = np.ones(np.shape(x)[0])
X = np.array([intercept, x]).T
beta_hat = np.linalg.inv(X.T @ X) @ X.T @ y

# Preparing the plot data for the estimated regression line
x_plot = np.linspace(-1, 1, 50)
y_est = beta_hat[0] + beta_hat[1] * x_plot

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x_plot, y_est, 'b--', lw=2, label='Estimate')
plt.plot(x_plot, true_beta[0] + true_beta[1] * x_plot, lw=2, c='g', label='Truth')
plt.plot(x, y, 'ro', markersize=4)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Least Squares Estimate vs. True Relationship')
plt.legend()
plt.grid(True)
plt.show(), beta_hat
```

c) How does the estimate of `beta` compare to the parameters we used to generate the data?

The estimate of \(\beta\) obtained from the least squares method should closely approximate the true parameters (\(1\) for the intercept and \(0.5\) for the slope) used to generate the data, assuming the noise is not overly large and the model correctly specifies the relationship. Variations between the estimated and true parameters are due to the randomness introduced by the noise in the data and the finite sample size.
